Import Libraries
These are essential Python libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional for better visualization aesthetics
sns.set(style='whitegrid')
```

Load the Dataset

```python
df = pd.read_csv(r"C:\Users\USER\Desktop\AI & ML INTERNSHIP\day2\Titanic-Dataset.csv") #reads the dataset.
df.head() #displays the first few rows.
```

 Basic Information & Null Values

info() shows data types and missing values.

isnull().sum() counts missing values per column.

```python
df.info()
df.isnull().sum()
```

Descriptive Statistics

Gives statistical overview: mean, std dev, min, max, percentiles â€” for numerical features only.

```python
df.describe()
```

Data Cleaning

```python
# Fill Age with median, drop Cabin, fill Embarked with mode
df['Age'].fillna(df['Age'].median())
df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop(columns='Cabin', inplace=True)
```

Univariate Analysis

1.Histogram & Boxplot

Helps spot outliers (boxplots) and shape/skew of distribution (histogram).


```python
numeric_features = df.select_dtypes(include=np.number).columns.tolist()
```

```python
for feature in numeric_features:
    plt.figure(figsize=(12,4))
    
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    
    plt.tight_layout()
    plt.show()
```

Categorical Features Analysis

```python
sns.barplot(x='Pclass', y='Fare', data=df)
plt.title("Average Fare by Passenger Class")
plt.show()
```

```python
categorical = df.select_dtypes(include='object').columns.tolist()

for feature in categorical:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[feature])
    plt.title(f'Countplot of {feature}')
    plt.xticks(rotation=45)
    plt.show()
```

Bivariate Analysis
ðŸ”¹ Survival Rate by Gender and Class


```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()

sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()
```

Correlation Analysis

Shows relationship between numerical variables.

Values close to +1 or -1 = strong correlation.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Pairplot for Detailed Relations

Visualizes scatter relationships, colored by survival.

```python
selected = ['Age', 'Fare', 'Pclass', 'Survived']
sns.pairplot(df[selected], hue='Survived')
```
